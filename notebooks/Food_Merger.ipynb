```python
import pandas as pd
import glob as gl
from IPython.display import display

all_files = gl.glob('/data/foodboost/*.csv', recursive=False)
```

```python
all_df = []
for f in all_files:
    if f.endswith("recipes.csv") :
        df =  pd.read_csv(f)
        df = df.drop(columns= df.columns[0], axis=1)
        df = df.rename(columns={'title': 'recipe'})
        #df.index = df.index.set_names(['ID'])
    else :
        df = pd.read_csv(f)
        df = df.drop(columns= df.columns[0], axis=1)
        df = df.groupby('recipe', sort=False).agg(lambda x: list(x))
        #df = df.reset_index()
        #df.index = df.index.set_names(['ID'])
    all_df.append(df)
```

```python
df = all_df[0]
for df_temp in all_df[1:]:
    df = pd.merge(df,df_temp, on='recipe', how='outer')
```
